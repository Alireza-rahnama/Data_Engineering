```python
#imports regularExpression to find a pattern(RegEx)
# The regular expression may neeed to be adjusted later to extract other data

import re

location=re.findall(r'\d+\°\ \d{2}\'\ \d+\.\d+ \w',output)
Latitude=[]
Latitude_Coord=[]
Longitude=[]
Longitude_Coord=[]
for entry in location:
    if entry[-1] == ('N' or 'S'):
        Latitude.append(entry)
    if entry[-1] == ('E' or 'W'):
        Longitude.append(entry)
print("Latitude",Latitude)
print("Longitude:",Longitude)
```

```python
#instantiated the list as plaeholder for the lattitude and longitude then transformed the string type to floating number to floating number
Latitude_Coord=[]
for i in Latitude:
    j=i.split(" ")
    k=float(j[2])
    # print("Latitude=",j)
    Latitude_Coord.append(k)
print("Latitude_Coord: ",Latitude_Coord)

Longitude_Coord=[]
for i in Longitude:
    j=i.split(" ")
    k=float(j[2])
    # print("Latitude=",j)
    Longitude_Coord.append(k)
print("Longitude_Coord: ", Longitude_Coord)
```

```python
# Create randomised data
import random
import pandas as pd
num_layers = 4
layer_names = ['Water','Mud', 'Silt', 'Limestone', 'Granite','Bottom']
layer_colours = ['cyan', 'grey', 'orange', 'red', 'green', 'black']

x = random.uniform(0, 10)
y = random.uniform(0, 10)
x_l = [x]*(num_layers+2)
y_l = [y]*(num_layers+2)

layer_depths = [0,3000]
for layer in range(0,num_layers):
    layer_depths.append(random.uniform(-3000,0))
layer_depths.sort()
layer_depths.reverse()


Well_Bore_Data = pd.DataFrame(data = {'x':x_l, 'y':y_l, 'layer_depths':layer_depths, 'layer_names':layer_names, 'layer_colours':layer_colours})
display(Well_Bore_Data)

# Create randomised data
```

```python
import plotly.graph_objects as go
z=[3000]*len(Longitude_Coord)

fig = go.Figure(data=[
     go.Scatter3d(x=Longitude_Coord, y=Latitude_Coord, z=z,
                  mode='markers',
                  marker=dict(size=8),
                  meta='diamond-open'
                  
                 ),
     go.Mesh3d(
        # 8 vertices of a cube
        alphahull=0,

        x=[0,0,50,50,0,0,50,50],
        y=[0,60,60,0,0,60,60,0],
        z=[0,0,0,0,3000,3000,3000,3000],
        flatshading = True,

        opacity=0.6,
        color='#FFFFCC'
    )                    
    ])



#Iteratively add layers of formation
for Layer in range(1, num_layers+1):
    shallow = Well_Bore_Data.iloc[Layer]['layer_depths']
    deep = Well_Bore_Data.iloc[Layer+1]['layer_depths']
    fig.add_trace(go.Mesh3d(
            alphahull=0,
        
            # 8 vertices of a rectangle
            x=np.array([0,0,50,50,0,0,50,50]),
            y=np.array([0,60,60,0,0,60,60,0]),
        
            z=np.array([shallow, shallow, shallow, shallow, deep, deep, deep, deep]),
              
            # Assign appriotiate identifiers to trace
            color = Well_Bore_Data.iloc[Layer]['layer_colours'],
            opacity = 0.5,
            name = Well_Bore_Data.iloc[Layer]['layer_names']

    ))

# Show Legend
fig.update_traces(showlegend=True)
fig.update_layout(scene = dict(
                    xaxis_title='Longitude_Coord',
                    yaxis_title='Latitude_Coord',
                    zaxis_title='TVD'),
                    width=1400,
                    height=700,
                    margin=dict(r=20, b=10, l=10, t=10))
fig.show()

# First checkpoint
# Generate base case for geological plot
# https://plotly.com/python/3d-mesh/
```
