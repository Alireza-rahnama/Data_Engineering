```python
def dataExtractor (directory): #Function which extracts data from ASC files given a directory argument for the ASC file
    columnsDataFrame = []
    columnsDataFrameTwo = []
    path = directory.replace("\\", '\\\\') #Replaces \\ in directory path with \\\\ so that python will recognize two slashes
    asciiFile = open(directory, 'r') #opens .ASC at directory argument
    columnsDataFrame = asciiFile.readlines()[11].strip().split() #strips and splits the 12th line of .ASC file, removing line endings and other unnecessary characters, and splitting separated data (by space) into list entries.
    asciiFile.close() #closes .ASC file, you must close and reopen it everytime you run a line of code to read data from the file.
    asciiFile = open(directory, 'r')
    columnsDataFrameTwo = asciiFile.readlines()[12].strip().split() #Strips and splits the 13th line of .ASC file
    asciiFile.close()
    i=0
    for entry in columnsDataFrame: #For each column header in columnsDataFrame list
        columnsDataFrame[i] = entry + " (" + columnsDataFrameTwo[i] + ")" #add the corresponding column header from columnsDataFrameTwo in brackets to the columnsDataFrame headers
        #This for loop creates column headers which specify units
        i +=1
    asciiFile = open(directory, 'r')
    rows = []
    numberOfLoop = 0
    for line in asciiFile.readlines():
        if numberOfLoop > 13: #skips all lines which do not contain data. This number may vary with other data structures.
            line = line.strip() #strips lines of line breaks and other unnecessary characters
            rows.append(line.split()) #appends data entries in .asc tables to the rows list
 
        
        numberOfLoop+=1
    df = pd.DataFrame(rows, columns=columnsDataFrame) #Creates pandas dataframe with row data provided by rows list, and corresponding column names provided by columnsDataFrame list entries
    asciiFile.close()
    directoryCreatePath = os.path.dirname(directory) #creates string directoryCreatePath with value being the folder path of the directory argument (removes file name and extention from string)
    directoryCreatePath = directoryCreatePath.replace("02.LWD_EWL", "ExtractedLWD_EWLASCData") #replaces 02.LWD_EWL in directory argument with ExtractedLWD_EWLASCData 
    if not os.path.exists(directoryCreatePath): #if output folder within ExtractedLWD_EWLASCData does not already exist, creates directory
        os.makedirs(directoryCreatePath)
    fileName = path.replace("02.LWD_EWL", "ExtractedLWD_EWLASCData") #replaces 02.LWD_EWL in path variable with ExtractedLWD_EWLASCData
    fileName = fileName.replace(".ASC", ".csv") #changes file extension in fileName string from .ASC to .csv for output to excel spreadsheet
    df.to_csv(fileName) #outputs pandas dataframe to csv file at path and filename specified in fileName string
    df = df[0:0] #resets pandas dataframe
```

```python
for i in glob.iglob('02.LWD_EWL/*/*.asc'): #enters 02.LWD_EWL folder which must be in the same directory as ASC Data Extraction.ipynb, enters each subfolder and each subfolder within this subfolder, and finds the path of each .ASC file stored there
    print (i, "beginning") #informs user that the process of data extraction for discovered file has begun
    if ('INF' in i or 'SUMMARY' in i or 'HEAD' in i or 'CON' in i or 'AUD' in i or 'info' in i): #checks if file path/name contains these strings - if they do, the data is not valuable.
        print(i, 'does not contain useful data.')
    else: 
        dataExtractor(i) #if the data is valuable, calls function dataExtractor with argument being the file path of the .ASC file
        print(i, "complete")
print("Extraction complete") #prints when all files have been discovered and have had data extracted to a .csv file or determined that their data is not valuable
```
